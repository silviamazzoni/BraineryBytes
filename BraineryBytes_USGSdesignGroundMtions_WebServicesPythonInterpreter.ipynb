```python
import requests as req
import json

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# https://github.com/jupyter-widgets/ipywidgets/issues/1853
from ipywidgets.widgets.interaction import show_inline_matplotlib_plots

import glob
import urllib
import webbrowser
from ipywidgets import widgets, Output

thisWidget = widgets.HTML(value = """<font color="#D47088"><h3>**********USGS WEBSERVICES Python Interpreter    by Silvia Mazzoni**********</h3></font>""")
display(thisWidget)
for ASCEcode in ['ASCE7-16','ASCE7-10','ASCE7-05','ASCE41-17','ASCE41-13','NEHRP-2020','NEHRP-2015','IBC-2015','IBC-2012','AASHTO-2009']:
#    print(ASCEcode)

    thisWidget = widgets.HTML(value = """<font color="#D47088"><h3>""" + ASCEcode + """</h3></font>""")
    display(thisWidget)

    ASCEcodeIn = ASCEcode.lower()
    thisURL = 'https://earthquake.usgs.gov/ws/designmaps/' + ASCEcodeIn + '.json?latitude=' + str(Latitude) +'&longitude=' + str(Longitude) +'&riskCategory=' + RiskCat + '&siteClass=' + SiteClass + '&title=' + label
    # these are the possible output spectra for all codes, not in all codes
    resp = req.get(thisURL)
    WEBresponse = json.loads(resp.text)

    spectraListAll = ['multiPeriodDesignSpectrum','multiPeriodMCErSpectrum','twoPeriodDesignSpectrum','twoPeriodMCErSpectrum','horizontalSpectrum','verticalDesignSpectrum','verticalMCErSpectrum','sadFloor','sadPercentileFactor','maxDirectionFactor']
    print("Request:")
    for label in WEBresponse['request'].keys():
        print(label + ': ' + str(WEBresponse['request'][label]))
    print("")

    if ASCEcode in ['ASCE41-17','ASCE41-13']:
        print("Response - Data:")
        for thisHazardLevelData in WEBresponse['response']['data']:
            print("")
            for label in thisHazardLevelData.keys():
                if not label in spectraListAll:
                    print(label + ': ' + str(thisHazardLevelData[label]))
            for thisSpectrum in spectraListAll:
                if thisSpectrum in thisHazardLevelData.keys():
                    thisArray = np.array(thisHazardLevelData[thisSpectrum])
                    x, y = thisArray.T
                    plt.plot(x,y,label=thisHazardLevelData['hazardLevel'])

    else:
        print("Response - Metadata:")
        for label in WEBresponse['response']['metadata'].keys():
            print(label + ': ' + str(WEBresponse['response']['metadata'][label]))
        print("")
        for label in WEBresponse['response']['data'].keys():
            if not label in spectraListAll:
                print(label + ': ' + str(WEBresponse['response']['data'][label]))

        if ASCEcode in ['NEHRP-2020']:
            for thisSpectrum in spectraListAll:
                if thisSpectrum in WEBresponse['response']['data'].keys():
                    thisArray = {}
                    thisArray = WEBresponse['response']['data'][thisSpectrum]
                    x = thisArray['periods']
                    y = thisArray['ordinates']
                    if len(x)>0:
                        plt.plot(x,y,label=thisSpectrum)
        else:
            for thisSpectrum in spectraListAll:
                if thisSpectrum in WEBresponse['response']['data'].keys():
                    thisArray = np.array(WEBresponse['response']['data'][thisSpectrum])
                    x, y = thisArray.T
                    plt.plot(x,y,label=thisSpectrum)

    plt.xlabel('Period (sec)')
    plt.ylabel('PSA (g)')
    plt.title(ASCEcode)
    plt.legend()
    plt.grid()
    plt.show()
```
